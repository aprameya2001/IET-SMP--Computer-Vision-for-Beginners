```python
import torch
import cv2
model = torch.hub.load('ultralytics/yolov3', 'yolov3') 
# Image
image=cv2.imread('test.jpeg')
# Inference
results = model(image)

```

```python

print(results.xyxy[0])
print(results)
results.print()  
results.show()
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```
