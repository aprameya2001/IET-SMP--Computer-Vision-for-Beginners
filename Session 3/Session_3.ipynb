## TOPIC 1- MORPHOLOGICAL TRANSFORMATIONS

- Morphological transformations are some simple operations based on the image shape. It is normally performed on binary images. 
- It needs two inputs, one is our original image, second one is called structuring element or kernel which decides the nature of operation. 
- Two basic morphological operators are Erosion and Dilation. 
- Then its variant forms like Opening, Closing, Gradient etc also comes into play.

**Types**

**1. Erosion**
- The basic idea of erosion is just like soil erosion only, it erodes away the boundaries of foreground object.
- The kernel slides through the image (as in 2D convolution). A pixel in the original image (either 1 or 0) will be considered 1 only if all the pixels under the kernel is 1, otherwise it is eroded (made to zero).
- So what happends is that, all the pixels near boundary will be discarded depending upon the size of kernel. So the thickness or size of the foreground object decreases or simply white region decreases in the image. 

```python
# EROSION

img = cv2.imread('./images/j.png', 0)
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)

print('TYPE 1: EROSION')
plt.subplot(1,2,1);
plt.title('Original Image')
plt.imshow(img, 'gray')

plt.subplot(1,2,2);
plt.title('Erosion')
plt.imshow(erosion, 'gray')
plt.show()
```

**2. Dilation**
- It is just opposite of erosion. 
- Here, a pixel element is '1' if atleast one pixel under the kernel is '1'. 
- So it increases the white region in the image or size of foreground object increases.

```python
# DILATION

img = cv2.imread('./images/j.png', 0)
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(img,kernel,iterations = 1)

print('TYPE 2: DILATION')
plt.subplot(1,2,1);
plt.title('Original Image')
plt.imshow(img, 'gray')

plt.subplot(1,2,2);
plt.title('Dilation')
plt.imshow(dilation, 'gray')
plt.show()
```

**3. Opening**
- Opening is just another name of **erosion followed by dilation**.

```python
# OPENING

img = cv2.imread('./images/opening.png', 0)
kernel = np.ones((5,5),np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

print('TYPE 3: OPENING')
plt.subplot(1,2,1);
plt.title('Original Image')
plt.imshow(img, 'gray')

plt.subplot(1,2,2);
plt.title('Opening')
plt.imshow(opening, 'gray')
plt.show()
```

**4. Closing**
- Closing is reverse of Opening, **Dilation followed by Erosion**.

```python
# CLOSING

img = cv2.imread('./images/closing.png', 0)
kernel = np.ones((5,5),np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

print('TYPE 4: CLOSING')
plt.subplot(1,2,1);
plt.title('Original Image')
plt.imshow(img, 'gray')

plt.subplot(1,2,2);
plt.title('Closing')
plt.imshow(closing, 'gray')
plt.show()
```

**5. Morphological Gradient**
- It is the **difference between dilation and erosion of an image**.
- The result will look like the outline of the object.

```python
# MORPHOLOGICAL GRADIENT

img = cv2.imread('./images/j.png', 0)
kernel = np.ones((5,5),np.uint8)
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)

print('TYPE 5: MORPHOLOGICAL GRADIENT')
plt.subplot(1,2,1);
plt.title('Original Image')
plt.imshow(img, 'gray')

plt.subplot(1,2,2);
plt.title('Morph Gradient')
plt.imshow(gradient, 'gray')
plt.show()
```

**6. Top Hat**
- It is the **difference between input image and Opening of the image**.

```python
# TOP HAT

img = cv2.imread('./images/j.png', 0)
kernel = np.ones((2,2),np.uint8)
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)

print('TYPE 6: TOP HAT')
plt.subplot(1,2,1);
plt.title('Original Image')
plt.imshow(img, 'gray')

plt.subplot(1,2,2);
plt.title('Top hat')
plt.imshow(tophat, 'gray')
plt.show()
```

## TOPIC 2- IMAGE SMOOTHING & BLURRING

- Image blurring is achieved by convolving the image with a low-pass filter kernel. 
- It is useful for removing noise. 
- It actually removes high frequency content (eg: noise, edges) from the image. 

### 1. 2D Convolution ( Image Filtering )
- OpenCV provides a function cv.filter2D() to convolve a kernel with an image

```python
img = cv2.imread('./images/opencv1.jpg')

kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)

plt.subplot(1,2,1);
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2);
plt.title('2D convolved Image')
plt.imshow(dst)
plt.show()
```

### 2. AVERAGE BLURRING

```python
img = cv2.imread('./images/opencv1.jpg')

blur = cv2.blur(img,(5,5))

plt.subplot(1,2,1);
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2);
plt.title('Average blurred Image')
plt.imshow(blur)
plt.show()
```

### 3. GAUSSIAN BLURRING
- Highly effective when image contains gaussian noise

```python
img = cv2.imread('./images/gaussian.png')

blur = cv2.GaussianBlur(img,(15,15),0)

plt.subplot(1,2,1);
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2);
plt.title('Gaussian Blurred Image')
plt.imshow(blur)
plt.show()
```

###  4. MEDIAN BLURRING
- Highly effective when the image contains salt and pepper noise

```python
img = cv2.imread('./images/salt_pepper_noise.jpg')

median = cv2.medianBlur(img,5)

plt.subplot(1,2,1);
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2);
plt.title('Median Blurred Image')
plt.imshow(median)
plt.show()
```

### 5. BILATERAL BLURRING
- Bilateral Blurring is highly effective in noise removal while keeping edges sharp.

```python
img = cv2.imread('./images/bilateral.jpg')

bilateral = cv2.bilateralFilter(img, 9, 75, 75)

plt.subplot(1,2,1);
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2);
plt.title('Bilateral blurred Image')
plt.imshow(bilateral)
plt.show()
```

## TOPIC 3- IMAGE GRADIENTS

- Formally, an image gradient is defined as a directional change in image intensity.
- The main application of image gradients lies within edge detection.
- Edge detection is the process of finding edges in an image, which reveals structural information regarding the objects in an image. 

### SOBEL & LAPLACIAN 

```python
#Resizing the Plots
plt.rcParams["figure.figsize"] = (20, 20)

img = cv2.imread('./images/sudoku.jpg', 0)

# 2nd parameter is a datatype to handle negative numbers
# 3rd parameter is kernel size. must be an odd integer
laplacian = cv2.Laplacian(img,cv2.CV_64F, ksize=3)

# sobelX gives the directional change in horizontal direction
# sobelY gives the directional change in vertical direction
# 2nd parameter is a datatype to handle negative numbers
# 3rd parameter is dx
# 4th parameter is dy
# 5th parameter is kernel size. must be an odd integer
sobelX = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
sobelY = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)

laplacian = np.uint8(np.absolute(laplacian))
sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))

sobelCombine = cv2.bitwise_or(sobelX, sobelY)

titles = ['image', 'Laplacian', 'sobelX', 'sobelY', 'sobel-combine']
images = [img, laplacian, sobelX, sobelY, sobelCombine]

for i in range(5):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])

plt.show()
```

## TOPIC 4- CANNY EDGE DETECTION

- A multistage algorithm
- Follows these steps:
    1. Noise Reduction
    2. Finding Intensity Gradient of the Image
    3. Non-maximum Suppression
    4. Hysteresis Thresholding

```python
# Example 1

img = cv2.imread('./images/sudoku.jpg', 0)

canny = cv2.Canny(img, 50, 80)
# Parameter 2 and 3 are threshold values

plt.subplot(1,2,1);
plt.title('Original Image')
plt.imshow(img, 'gray')

plt.subplot(1,2,2);
plt.title('Canny edge detection')
plt.imshow(canny, 'gray')
plt.show()
```
